```python
from sklearn.tree import DecisionTreeRegressor
```

```python
import numpy as np
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
d_man = 25 + np.random.random(30)
d_dog = 12 + np.random.random(30)
d_cat = 12 + np.random.random(30)
d_woman = 25 + np.random.random(30)
```

```python
import pandas as pd
```

```python
d_man = pd.concat([pd.DataFrame(d_man),pd.DataFrame(["Man"]*30)],axis = 1 )
```

```python
d_dog = pd.concat([pd.DataFrame(d_dog),pd.DataFrame(["Dog"]*30)],axis = 1 )
```

```python
d_cat = pd.concat([pd.DataFrame(d_cat),pd.DataFrame(["Cat"]*30)],axis = 1 )
```

```python
d_woman = pd.concat([pd.DataFrame(d_woman),pd.DataFrame(["Woman"]*30)],axis = 1 )
```

```python
df = pd.concat([d_man,d_dog,d_woman,d_cat], axis = 0).reset_index()
```

```python
df = df.drop("index",axis = 1)
df.columns = ["Height","Tips"]
```

```python
df_later = df.copy()
```

```python
listo = ["Man","Dog","Woman","Cat"]
```

```python
k = LabelEncoder()
df.Tips = k.fit(listo).transform(df.Tips)
```

```python
the_tree = DecisionTreeRegressor(criterion='mse' , splitter = 'best' , max_depth = 1)
```

```python
#Cat = 0  , Dog = 1, Woman = 3 , Man = 2
%matplotlib inline 
#Easy split at 1.5
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(1)
plt.scatter(df.Tips,df.Height)
```

```python
the_tree.fit(np.expand_dims(df.Tips,axis =1 ),df.Height)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(the_tree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
saved_tree = DecisionTreeRegressor(criterion='mse' , splitter = 'best' , max_depth = 2 , max_leaf_nodes=20 , min_impurity_decrease= 0)
```

```python
saved_tree.fit(np.expand_dims(mergo.Encoding,1),mergo.Height)
```

# **Mean Encoding to the rescue** (Sorts according to target mean in each category for easy seperation by trees)

```python
df_later.head()
```

```python
df_later.Tips = df_later.Tips.astype(str)
mean_encoder.Tips = mean_encoder.Tips.astype(str)
```

```python
dummy = df_later.groupby("Tips")["Height"].sum().reset_index()
```

```python
mean_encoder = df_later.groupby("Tips")["Height"].mean().reset_index()
```

```python
mean_encoder.columns = ["Tips", "Tips_Encoded"]
```

```python
#But when I do mean_encoder.columns.values[-1] = "Tips_Encoded"    ,,,, mean_encoder["Tips_Encoded"] throws an error ,
#also merging thros nan
```

```python
merg2 = pd.merge(df_later, mean_encoder , left_on = "Tips",right_on = "Tips" , how="left")
```

```python
smart_tree = DecisionTreeRegressor(criterion='mse' , splitter = 'best' , max_depth = 1 , max_leaf_nodes=2 , min_impurity_decrease= 0)
```

```python
smart_tree.fit(np.expand_dims(merg2.Tips_Encoded,1),merg2.Height)
```
