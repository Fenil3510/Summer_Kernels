```python
def eval_error(pred,dtrain):
    labels = dtrain.get_label()
    pred[pred > 20] = 20
    pred[pred < 0] = 0
    labels[labels > 20] = 20
    labels[labels < 0] = 0
    return 'eval_error' , np.sqrt(sum((pred - labels)**2)/len(pred))
```
